# Tutorial 3 Pandas : Data Formatting (merge, sort, reset_index, fillna)

```python
import pandas as pd
import numpy as np
```

```python
file = './Data/data.csv'
file2 = './Data/data_makanan.csv'
zoo = pd.read_csv(file, delimiter = ',')
zoo_eats = pd.read_csv(file2, delimiter = ';')
```

### Merge Data

```python
zoo.merge(zoo_eats)
```

```python
zoo_eats.merge(zoo)
```

<img src="https://data36.com/wp-content/uploads/2018/08/4-pandas-merge-inner-outer-left-right-768x579.png">

```python
zoo.merge(zoo_eats, how = 'outer')
```

```python
zoo.merge(zoo_eats, how = 'left')
```

```python
zoo.merge(zoo_eats, how = 'right')
```

#### <font color=red>Penting</color>

##### Terkadang ada mismatch dalam nama kolom sehingga dapat dilakukan code seperti dibawah ini

```python
zoo.merge(zoo_eats, how = 'left', left_on = 'animal', right_on = 'animal')
```

### Sortng Data

```python
zoo.sort_values('water_need')
```

```python
zoo.sort_values(by = ['animal', 'water_need'])
```

```python
zoo.sort_values(by = ['water_need'], ascending = False)
```

### Reset Index reset_index()

```python
zoo.sort_values(by = ['water_need'], ascending = False).reset_index()
```

```python
zoo.sort_values(by = ['water_need'], ascending = False).reset_index(drop = True)
```

### Fill+NA fillna

```python
zoo.merge(zoo_eats, how = 'left').fillna('unknown')
```

```python
zoo.merge(zoo_eats, how = 'left').fillna('meat')
```
